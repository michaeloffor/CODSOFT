```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import time
print('All libraries imported')
```

```python
st = time.time()

# loading the dataset
data = 'IMDb Movies India.csv'
df_movie = pd.read_csv(data)
df_movie.head(20)
```

```python
#getting the number of rows and column in the dataset
df_movie.shape
```

```python
#to describe the dataset
df_movie.describe(include='all')
```

```python
#getting info about the whole dataset
df_movie.info()
```

```python
#getting the total number of null values in each column
df_movie.isnull().sum()
```

```python
#getting the maximum and minimum rating
df_movie['Rating'].describe()[['max', 'min']]
```

```python
#firstly to process the ratings column, i need to boxplot to check for ouyliers
sns.boxplot(df_movie['Rating'])
```

```python
#we see that there is a significant outliers, so we will use median, but we won't use a single value, rather range that is close to the median
# Generate random floats within a range close to the median
lower_bound = df_movie['Rating'].median() - 0.5
upper_bound = df_movie['Rating'].median() + 0.5

# Fill the null values in the 'Rating' column with random floats and verify if all has been filled
df_movie['Rating'] = df_movie['Rating'].apply(lambda x: np.random.uniform(lower_bound, upper_bound) if pd.isnull(x) else x)
df_movie.isnull().sum()
df_movie['Rating']
```

```python
# Fill the null values in the 'Genre' column with the mode
mode_genre = df_movie['Genre'].mode()[0]
df_movie['Genre'] = df_movie['Genre'].fillna(mode_genre)

# Verify if all null values have been filled
df_movie.isnull().sum()
```

```python
# Fill the null values in the 'Director' column with the mode
mode_director = df_movie['Director'].mode()[0]
df_movie['Director'] = df_movie['Director'].fillna(mode_director)

# Verify if all null values have been filled
df_movie.isnull().sum()
```

```python
# Fill the null values in the 'Actor 1' column with the mode
mode_actor1 = df_movie['Actor 1'].mode()[0]
df_movie['Actor 1'] = df_movie['Actor 1'].fillna(mode_actor1)

# Fill the null values in the 'Actor 2' column with the mode
mode_actor2 = df_movie['Actor 2'].mode()[0]
df_movie['Actor 2'] = df_movie['Actor 2'].fillna(mode_actor2)

# Fill the null values in the 'Actor 3' column with the mode
mode_actor3 = df_movie['Actor 3'].mode()[0]
df_movie['Actor 3'] = df_movie['Actor 3'].fillna(mode_actor3)

# Verify if all null values have been filled
df_movie.isnull().sum()
```

```python
#to fill in the null values in year, duration and votes
#first modify the data and set it to the appropriate datatype

#for year, we need to first set it as float, fill in the null value and then set as integer
df_movie['Year'] = df_movie['Year'].str.extract(r'\((\d{4})\)').astype(float)

#for duration
df_movie['Duration'] = df_movie['Duration'].str.replace(' min', '').astype(float)

#for votes, there some unwanted string in the column that needs to be cleared
# Clean the 'Votes' column with a custom function
def clean_votes(vote):
    if isinstance(vote, str):
        vote = vote.replace('$', '').replace('M', 'e6').replace('K', 'e3')
        try:
            return float(vote)
        except ValueError:
            return np.nan
    return vote

try:
    df_movie['Votes'] = df_movie['Votes'].apply(clean_votes)
except Exception as e:
    print(f"Error cleaning 'Votes' column: {e}")
```

```python
#now, we have to change the various datatypes to its correct form but lets first find out the current form of the column 
df_movie.dtypes
```

```python
df_movie.head(20)
```

```python
#boxplotting to check for outliers in the year column
sns.boxplot(df_movie.Year)
```

```python
#no significant outlier, therefore i will use mean
mean = df_movie['Year'].mean()
df_movie['Year'] = df_movie['Year'].fillna(mean)
df_movie.isnull().sum()
```

```python
#now that the year column has been filled, it can now be changed to int
df_movie['Year'] = df_movie['Year'].astype(int)
df_movie.head(25)
```

```python
#boxplotting for duration
sns.boxplot(df_movie.Duration)
```

```python
#there are outliers, but they are not extreme, so i will use values close to median
median = df_movie['Duration'].median()
value = np.random.randint(median-3, median+3)
df_movie['Duration'] = df_movie['Duration'].fillna(value)
df_movie.isnull().sum()
```

```python
df_movie.head(25)
```

```python
#now tha we have filled the null value in Duration, lets convert it to integer
df_movie['Duration'] = df_movie['Duration'].astype(int)
df_movie.head(25)
```

```python
#to boxplot votes

sns.boxplot(df_movie.Votes)
```

```python
#seeing the extreme outliers, we need to use another method to fill the column
from scipy.stats import mstats

# Winsorize the 'Votes' column to cap the extreme values at 1% and 99%
df_movie['Votes'] = mstats.winsorize(df_movie['Votes'], limits=[0.01, 0.01])

# Fill the remaining null values with the median
median_votes = df_movie['Votes'].median()
df_movie['Votes'] = df_movie['Votes'].fillna(median_votes)

# Verify if all null values have been filled
df_movie.isnull().sum()
```

```python
#mow that we have filled the votes column, lets convert it to integer
df_movie['Votes'] = df_movie['Votes'].astype(int)
df_movie.head(25)
```

```python

#The Genre column contain more than one value so we have to split the 'Genre' column into multiple rows
df_movie = df_movie.assign(Genre=df_movie['Genre'].str.split(',')).explode('Genre')
df_movie.head(25)
```

```python
#drop column that might introduce noise
df_movie.drop(columns=['Name','Year', 'Votes'], inplace=True)
df_movie.head(25)
```

```python
#converting categorical variable to numerical variable
from sklearn.preprocessing import LabelEncoder
Director = LabelEncoder().fit_transform(df_movie['Director'])
Actor1 = LabelEncoder().fit_transform(df_movie['Actor 1'])
Actor2 = LabelEncoder().fit_transform(df_movie['Actor 2'])
Actor3 = LabelEncoder().fit_transform(df_movie['Actor 3'])
Genre = LabelEncoder().fit_transform(df_movie['Genre'])
df_movie['Director'] = Director
df_movie['Actor 1'] = Actor1
df_movie['Actor 2'] = Actor2
df_movie['Actor 3'] = Actor3
df_movie['Genre'] = Genre
df_movie.head(25)
```

```python
#splitiing the data into feature and target as x and y respectively

x = df_movie.drop('Rating', axis=1)
y = df_movie['Rating']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Split the dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Create a Linear Regression model
regr = LinearRegression()

# Train the model using the training sets
regr.fit(x_train, y_train)

# Make predictions using the testing set
y_pred = regr.predict(x_test)

# Calculate and print the Mean Absolute Error (MAE)
mae = metrics.mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error:', mae)

# Calculate and print the Mean Squared Error (MSE)
mse = metrics.mean_squared_error(y_test, y_pred)
print('Mean Squared Error:', mse)

# Calculate and print the Root Mean Squared Error (RMSE)
rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', rmse)
```

```python
et = time.time()
elp = et - st
print('The time taken for the process is ', round(elp, 2))
```

```python
# Input new data
new_data = (3, 1, 34.50, 0, 0, 7.8292)
new_array = np.asarray(new_data)

# Reshape the array
new_array_reshape = new_array.reshape(1, -1)

# Make predictions using the trained model
new_prediction = regr.predict(new_array_reshape)

# Set a threshold for high rating
threshold = 7.5

# Determine if the prediction is high or low
if new_prediction > threshold:
    prediction_result = 'High'
else:
    prediction_result = 'Low'

# Print the prediction result
print(f"The predicted rating is: {prediction_result}")
```
